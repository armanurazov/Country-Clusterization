# Exploratory Data Analysis

Check if we have any missing data

```python
df.isnull().sum()
```

```python
df[df['Agriculture'].isnull()]['Country']
```

```python
#fill N/A with 0
df[df['Agriculture'].isnull()] = df[df['Agriculture'].isnull()].fillna(0)
```

```python
df.isnull().sum()
```

```python
#fill with region mean
df['Climate'] = df['Climate'].fillna(df.groupby('Region')['Climate'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
df[df['Literacy (%)'].isnull()]
```

```python
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

```python
df = df.dropna()
```

# Data Preparation

column Country has a unique value, therefore do not hold any value for clusterisation

```python
X = df.drop("Country",axis=1)
```

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

# Data Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

# K means clustering model 

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,30):
    
    model = KMeans(n_clusters=k)   
    model.fit(scaled_X)  
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```

# Model interpretation

let's take example of K = 3

```python
model = KMeans(n_clusters=3)
model.fit(scaled_X)
```

```python
model.labels_
```

```python
X['K=3 Clusters'] = model.labels_
```

```python
X.corr()['K=3 Clusters'].sort_values().plot(kind='bar')
```
